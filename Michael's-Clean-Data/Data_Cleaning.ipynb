```python
# Import Dependencies
import pandas as pd
```

```python
# Import the games file
games_file = "./Resources/games.csv"

# Load games data into a dataframe
games_df = pd.read_csv(games_file)

games_df.head()
```

```python
# Count the data
games_df.count()
```

```python
# Drop NA values
games_df = games_df.dropna(how='any')
games_df.count()
```

```python
# Check for duplicates
duplicates = games_df[games_df.duplicated()]
duplicates
```

```python
# Drop duplicates
games_df.drop_duplicates(keep='first', inplace=True)
games_df.count()
```

```python
# Now check the game details file
# Create a variable to hold the file path
game_details = './Resources/games_details.csv'

# Import data into a dataframe
gd_df = pd.read_csv(game_details)

# Preview the data
gd_df.head()
```

```python
# Count the data
gd_df.count()
```

```python
# Drop NA values
gd_df = gd_df.dropna(how='any')

# Re-count the data
gd_df.count()
```

```python
# Now merge the data
merge = pd.merge(games_df, gd_df, on='GAME_ID', how='inner')

# Preview the data
merge.head()
```

```python
# Count the data
merge.count()
```

```python
# Preview the dataframe 
merge.head()
```

```python
# Check the end of the dataframe
merge.tail()
```

```python
# Set the Game Date as the index
merge = merge.set_index("GAME_DATE_EST")

# Preview the data
merge.head()
```

```python
# Now look again at the tail
merge.tail()
```

```python
merge
```

```python
# Reset the index
merge = merge.reset_index()

# Preview the data
merge.head()
```
